```python
import pandas as pd

df=pd.read_csv('/content/Creditcard_data.csv')

df.head()
```

```python
df['Class'].value_counts()
```

```python
selected_columns=['V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28']
X=df[selected_columns]
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
from collections import Counter

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.35,random_state=100)
print(f"Training target statistics: {Counter(y_train)}")
print(f"Testing target statistics: {Counter(y_test)}")

```

```python
#oversampling
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(sampling_strategy='minority')
X_resampled, y_resampled = oversampler.fit_resample(X_train, y_train)
```

```python
#undersampling
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(sampling_strategy='majority')
X_resampled, y_resampled = undersampler.fit_resample(X_train, y_train)
```

```python
# Merging the Balanced Data
train_data = pd.concat([X_resampled, y_resampled], axis = 1)
print(train_data)
```

```python
pd.Series(y_resampled).value_counts()
```

```python
#random sampling
import math
import pandas as pd
import random
import numpy as np
# Given values
Z = 1.96
p = 0.5
E = 0.05

sample_size= math.floor((Z**2 * p * (1 - p)) / E**2)


print("Sample Size:", sample_size)

data=pd.read_csv('/content/Creditcard_data.csv')

random_indices = random.sample(range(len(data)), sample_size)
sample = data.iloc[random_indices]
print(sample)

np.random.seed=42

```

```python
ed = train_data.sample(n = sample_size, random_state = 42)

x = ed.drop('Class', axis=1)
y = ed['Class']
ed_1_x_train, ed_1_x_test, ed_1_y_train, ed_1_y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Cluster Sampling
def get_clustered_Sample(df, n_per_cluster, num_select_clusters):
    N = len(df)
    K = int(N/n_per_cluster)
    data = None
    for k in range(K):
        sample_k = df.sample(n_per_cluster)
        sample_k["cluster"] = np.repeat(k,len(sample_k))
        df = df.drop(index = sample_k.index)
        data = pd.concat([data,sample_k],axis = 0)

    random_chosen_clusters = np.random.randint(0,K,size = num_select_clusters)
    samples = data[data.cluster.isin(random_chosen_clusters)]
    return(samples)

ed_3 = get_clustered_Sample(df = train_data, n_per_cluster = 100, num_select_clusters = 2)
ed_3
```
